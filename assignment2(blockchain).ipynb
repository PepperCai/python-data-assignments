```python
import pandas as pd
from matplotlib import pyplot as plt
df = pd.read_csv('blockchain.csv') #read csv
#df.head()
```

```python
#看看点赞和喜欢的作者，哪些在前面，并且有个排序
pd_df2 = df.groupby(['author'])['thumb'].mean().reset_index().sort_values("thumb",ascending=False)
pd_df3 = df.groupby(['author'])['favorite'].mean().reset_index().sort_values("favorite",ascending=False)#作何byfavorite和thumb
attr = pd_df2.author
v1 = pd_df2.thumb  #you can pass a list like data here
v2 = pd_df3.favorite
bar = Bar("author by thumb and favorite")
bar.add("by thumb", attr, v1, mark_line=["average"])
bar.add("by favorite", attr, v2, mark_line=["average"])
bar
```

```python
plt.figure(figsize=(6,9))#参照helper饼状图做法
labels = [u'蓝狐笔记',u'亿欧网',u'清风徐来028',u'人人都是产品经理',u'Anne']
sizes = [50,30,7,6,5]
colors = ['#F0E68C','#40b4e5','#FF7F50','#E0FFFF','#F5F5DC']
explode = (0.01,0,0,0,0)
patches,l_text,p_text = plt.pie(sizes,explode=explode,labels=labels,colors=colors,
                                labeldistance = 1.1,autopct = '%3.1f%%',shadow = False,
                                startangle = 90,pctdistance = 0.6)
plt.axis('equal')
plt.legend()
plt.show()
```

```python
plt.rcParams['font.sans-serif']=['SimHei'] #set for displaying Chinese characters here.
plt.rcParams['axes.unicode_minus']=False #set for displaying `-`


author_counts=df['author'].value_counts()[:5].sort_values(ascending=False) #sort values
author = pd.DataFrame(author_counts)
fig = plt.figure(figsize=(14,7)) #adjust size
plt.bar(author.index, author.author,color = '#F0E68C',edgecolor = '#F0E68C') #change color of the bars

plt.title('top 5 author') #plot title and label name
plt.xlabel('author')
plt.ylabel('author')
plt.show()
```

```python
from pyecharts import Bar,Line,Overlap

attr=[u'清风徐来028',u'亿欧网',u'蓝狐笔记',u'人人都是产品经理',u'Anne']
v1=[109, 77, 64, 42.7, 42]
v2=[20, 11, 11, 10, 7]
bar=Bar('最受欢迎的区块链作者的喜欢和赞')
bar.add('喜欢',attr,v1)
line=Line()
line.add("赞",attr,v2)

overlap=Overlap()
overlap.add(bar)
overlap.add(line)

overlap
```
